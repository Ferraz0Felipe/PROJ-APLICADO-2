Importação do Data Frame

```python
import requests
import pandas as pd
from io import StringIO

# Carregando arquivo
url = "https://raw.githubusercontent.com/Ferraz0Felipe/PROJ-APLIC-I/main/Dados/Dados_alternadores_garantia.csv"

# Faz o download do arquivo CSV
response = requests.get(url)

# Verifica se o download foi bem-sucedido
if response.status_code == 200:

    # Lê o conteúdo do arquivo CSV usando o pandas
    csv_data = response.content.decode('utf-8')

    # Converte o conteúdo para um DataFrame do pandas
    df = pd.read_csv(StringIO(csv_data), sep=";")
    print(df.head())

else:
    print('Falha ao baixar o arquivo CSV.')

# Mostrar informações

linhas = csv_data.split("\n")
```

Criando um novo DF para Aprendizado de Maquina

```python
novo_df = df[['Customer', 'KMS', 'Technical Factor', 'MIS', 'KM/Year']]
```

```python
novo_df.head()
```

Transformando os dados de KM/Yar em Int64

```python

novo_df['KM/Year'] = novo_df['KM/Year'].str.replace(',', '.').astype(float)

novo_df['KM/Year'] = novo_df['KM/Year'].astype('int64')
```

```python
novo_df.dtypes
```

Importando Pacotes que serão utilizados no treino 

```python
import pandas as pd
import requests
from io import StringIO
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.metrics import classification_report
```

Trantando e criando treinamento de maquina com K Vizinho 3

```python
# Separar os dados de entrada (X) e saída (y)
X = novo_df.drop(columns=['Customer'])

# Normalizar os dados de entrada (X)
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)

y = novo_df['Customer']

# Separar os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123, stratify=y)

#definição

clf = KNeighborsClassifier(3)

# Treinamento 

clf.fit(X_train,y_train)

#Avaliação

y_pred = clf.predict(X_test)

print(clf.score(X_test,y_test))

print('\nClassification Report:\n')
print(classification_report(y_test,y_pred))
```

Resultado para novos 5000 casos

```python
import numpy as np

# Gerar novos dados de entrada (5000 registros, com base na distribuição original)
num_novos_registros = 5000
X_novos = np.random.rand(num_novos_registros, X_train.shape[1])

# Normalizar os novos dados
X_novos = scaler.transform(X_novos)

# Prever os novos valores de Customer
y_novos_pred = clf.predict(X_novos)

# Contar a distribuição dos valores previstos de Customer
distribuicao_novos_customers = pd.Series(y_novos_pred).value_counts()

print(distribuicao_novos_customers)
```
